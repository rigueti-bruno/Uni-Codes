#### Junção por Chaves Estrangeiras com o metodo merge:

```python
df1 = pd.DataFrame({'lkey': ['foo', 'bar', 'baz', 'foo'],'value': [1, 2, 3, 5]})
df2 = pd.DataFrame({'rkey': ['foo', 'bar', 'baz', 'foo'],'value': [5, 6, 7, 8]})
print(df1.merge(df2, left_on='lkey', right_on='rkey', suffixes=('_left','_right')))
```

#### Junção pelos index padrão com o metodo join:

```python
df3=pd.DataFrame({'key':['K0','K1','K2','K3','K4','K5'],'A':['A0','A1','A2','A3','A4','A5']})
other=pd.DataFrame({'key':['K0','K1','K2'],'B':['B0','B1','B2']})
df3.join(other,lsuffix='_caller',rsuffix='_other')
```

#### Agrupando Dados com o metodo groupby:

```python
df5=pd.DataFrame({'Animal':['Falcon','Falcon','Parrot','Parrot'],'Max Speed':[380.,370.,24.,26.]})
grouped=df5.groupby(['Animal'])
print(grouped.mean()) # exibe a agregação da média para os grupos de animais classificados
```

#### Extraindo apenas duas colunas do DataFrame carros com o método filter utilizando seus labels:

```python
vendasano = carros.filter(['year','sellingprice']) # extrai a coluna dos anos e a coluna dos preços de venda e as atribui a um novo DataFrame
```

#### Calculando uma série de agregações em relação às vendas anuais de carros:

```python
agrupado = vendasano.groupby('year')
agregar=agrupado.agg(['mean','sum','count'])
agregar.reset_index(inplace=True)
agregar.tail(5)
```

#### Utilizando o metodo apply para executar funções com os valores de um DataFrame:

```python
df6=pd.DataFrame([[4,9]]*3,columns=['A','B']) # cria o DataFrame
df6
```

```python
df6.apply(np.sqrt) # aplica a função raiz quadrada a cada elemento do DataFrame
```

```python
df6['A']=df6['A'].apply(lambda x: x + 1) # soma com uma função lambda 1 a cada elemento da coluna A
df6
```

```python
df6['C']=df6.apply(lambda x: x['A']+x['B'],axis=1) # soma as colunas A e B e atribui o resultado a uma nova coluna C
df6
```

```python
df6['C']=df6['C'].apply(lambda x: x*x)
df6
```
