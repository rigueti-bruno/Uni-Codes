Carregando uma Planilha:

```python
import xlrd # carrega a biblioteca de manipulação de planilhas
```

```python
planilha = xlrd.open_workbook("dicionario_pessoas.xls") # carrega a pasta de trabalho do Excel para o código
```

```python
primeira_aba = planilha.sheet_by_index(0) # seleciona a primeira planilha da Pasta de Trabalho do Excel
```

```python
# Testes:
print("Nome:", primeira_aba.name) # imprime o nome da planilha
print("Num linhas:", primeira_aba.nrows) # imprime o numero de linhas da planilha
print("Num colunas:", primeira_aba.ncols) # imprime o numero de colunas da planilha
```

```python
# Imprime os conteúdos de cada linha da planilha:
for idx, linha in enumerate(primeira_aba.get_rows()):
    print(linha)
```

```python
# Cria uma Classe que vai armazenar as Classes da Planilha:
class Variavel:
    # Construtor:
    def __init__(self,posicao_inicial,tamanho,codigo,descricao):
        self.posicao_inicial = posicao_inicial
        self.tamanho = tamanho
        self.codigo = codigo
        self.descricao = descricao
        self.categoria = []
    
    #Cria uma lista de categorias:
    def add_categoria(self,categoria):
        #categoria = dict({'tipo':tipo,'descricao':descricao})
        self.categoria.append(categoria)
    
    #
    def __str__(self):
        return self.codigo + ".." + self.descricao
```

```python
# Extrai os dados da planilha para criar os objetos das Categorias:
variaveis = [] # armazenara uma lista com as variaveis para o construtor
nova_variavel = None # conterá as variaveis que serão criadas para armazenar no construtor

for idx, linha in enumerate(primeira_aba.get_rows()): # carrega os dados das linhas da planilha
    print(linha)
    primeira_celula = linha[0]
    if primeira_celula.ctype == 2:
        if nova_variavel:
            variaveis.append(nova_variavel)
        #Cria um objeto da classe Variavel:
        posicao_inicial = linha[0].value
        tamanho = linha[1].value
        codigo = linha[2].value
        descricao = linha[4].value
        nova_variavel = Variavel(posicao_inicial,tamanho,codigo,descricao)
        
        #Adiciona uma categoria ao objeto nova_variavel:
        categoria_tipo = linha[5].value
        categoria_descricao_tipo = linha[6].value
        nova_variavel.add_categoria({'categoria_tipo':categoria_tipo,'categoria_descricao_tipo':categoria_descricao_tipo})
        
    else:
        if nova_variavel:
            categoria_tipo = linha[5].value
            categoria_descricao_tipo = linha[6].value
            nova_variavel.add_categoria({'categoria_tipo':categoria_tipo,'categoria_descricao_tipo':categoria_descricao_tipo})
    
    #if idx > 50:
        #break
```

```python
# Conferencia dos valores extraidos:

# Imprime o numero de variaveis extraidas:
print('Total de variaveis:',len(variaveis))

# Imprime os dados das categorias das variaveis extraidas:
for variavel in variaveis:
    print(variavel)
    for categ in variavel.categoria:
        print('\t',categ)
```
