### Decomposição de uma Série Temporal no Python

```python
# Importando as Bibliotecas necessárias:
import pandas as pd
from statsmodels.tsa.seasonal import seasonal_decompose
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Carregando os Dados em um Dataframe:
df = pd.read_excel('planilha_dados_1.xlsx')
```

```python
# Conferir as 5 primeiras entradas do dataframe:
df.head(5)
```

```python
# Conferir as informações sobre os dados:
df.info()
```

```python
# Plotar um gráfico da Produção Industrial mensal PIM-PF:
plt.plot(df.Month, df.Value)
```

```python
# Antes de fazer a decomposição da série, temos que transformar as datas em índice:
pd.Month = pd.to_datetime(df.Month)
df.set_index('Month', inplace=True)
```

```python
# Para fazer a decomposição da série de acordo com o Modelo Aditivo:
result = seasonal_decompose(df,model='additive').plot()
```

```python
# Para fazer a decomposição da série de acordo com o Modelo Multiplicativo:
result = seasonal_decompose(df,model='multiplicative').plot()
```
