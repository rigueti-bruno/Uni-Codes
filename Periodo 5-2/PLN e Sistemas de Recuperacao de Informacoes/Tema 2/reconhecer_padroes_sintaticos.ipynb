Vamos Identificar o Padrão Sintático em um Texto:

```python
# Importando as bibliotecas necessárias:
import re
from nltk.tokenize import word_tokenize
```

```python
# Definindo o texto a ser analisado:
c_codigo = '''
#include <stdio.h>

int main() {
    int num = 5;
    if (num % 2 ==0) {
        printf("Número Par");
    }
    return 0;
}'''
```

```python
# Tekenizando o texto:
tokens = word_tokenize(c_codigo)
```

```python
# Definindo a regra para reconhecer padrões sintáticos:
padrao_sintatico = r"if\s*\([^)]+\)\s*\{[^}]*\}"
```

```python
# Buscando o padrão sintático no texto tokenizado e exibindo o resultado:
sintaxe_if = re.findall(padrao_sintatico, ' '.join(tokens))
for s in sintaxe_if:
    print(s)
```

```python
# Definindo um novo texto:
c_codigo_2 = '''
#include <stdio.h>

int main() {
    int num = 1;
    while (num <= 10) {
        printf("%d ", num);
        num++;
    }
    return 0;
} '''
```

```python
# Tokenizando o novo texto:
tokens = word_tokenize(c_codigo_2)
```

```python
# Definindo uma nova regra de validação:
padrao = r"while\s*\([^)]+\)\s*\{[^}]*\}"
```

```python
# Executando a análise e imprimindo o resultado:
comando_while = re.findall(padrao, ' '.join(tokens))
for s in comando_while:
    print(s)
```
