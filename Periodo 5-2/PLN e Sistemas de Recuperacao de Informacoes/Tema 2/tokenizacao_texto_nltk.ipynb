ToKenização de Texto com a Biblioteca NLTK do Python

```python
# Importando a biblioteca nltk:
import nltk
# Baixando os recursos necessários do nltk:
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
# Tokenizando um texto com a NLTK:
from nltk.tokenize import word_tokenize # importando a função de tokenização de palavras
texto = "Eu estudo processamento de linguagem natural." # texto a ser tokenizado
tokens = word_tokenize(texto) # tokenizando o texto
print(tokens) # imprimindo os tokens
```

Filtrando e imprimindo palavras do texto tokenizado:

```python
# Filtrando palavras do texto tokenizado:
palavras_filtradas = [palavra for palavra in tokens if len(palavra) > 4]

# Imprimindo as palavras filtradas:
for p in palavras_filtradas:
    print(p)
```
