Vamos eliminar algumas palavras selecionadas de um texto com regex.

```python
# Importando a biblioteca Regex:
import re
```

```python
# Indicando o texto a ser limpo:
texto = "Este texto é um exemplo para fazer um checklist de revisão do motor do carro."
```

```python
# Indicação das palavras a serem removidas do texto:
palavras_para_eliminar = ['este','para','de']
```

```python
# Colocando todo texto em letras minusculas:
texto = texto.lower()
```

```python
# Eliminando as palavras com Regex:
for palavra in palavras_para_eliminar:
    padrao = r'\b%s\b' % re.escape(palavra)
    texto = re.sub(padrao,"",texto)
```

```python
# Exibindo o texto após ser limpo:
print(f'O resultado é: {texto}')
```
