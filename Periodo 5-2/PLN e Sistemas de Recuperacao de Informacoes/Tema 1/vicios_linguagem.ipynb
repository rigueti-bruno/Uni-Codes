Vamos utilizar a biblioteca regex para verificar se há no texto paralavras consideradas vícios de linguagem.

```python
# Importando a biblioteca regex:
import re
```

```python
# Indicação das palavras que representam vícios de linguagem:
vicios = ["né","enfim"]
```

```python
# Texto simples a ser utilizado:
texto = 'Este é um texto, né? Temos que encontrar os vícios de linguagem, enfim.'
```

```python
# Localizando os vicios de linguagem no texto utlizando a função search da biblioteca regex:
palavras_encontradas = []

for padrao in vicios:
    if re.search(padrao, texto):
        palavras_encontradas.append(padrao)
```

```python
# Exibindo a conclusão sobre os vícios de linguagem encontrados:
print(f'O total de vicios de linguagem encontrados no texto é: {len(palavras_encontradas)}')
```
