Vamos extrair datas de um texto com Regex

```python
# Importando a biblioteca regex:
import re
```

```python
# Criando um texto e atribuindo a uma variavel:
texto = "Existem várias datas comemorativas, como o natal no dia 25/12 e 1/1"
```

```python
# Criando um padrão de data a ser extraida do texto com Regex:
padrao = r'\b(\d{1,2})/(\d{1,2})\b'
```

```python
# Econtrando as datas no texto de acordo com o padrão definido:
combinacoes = re.findall(padrao, texto)
```

```python
# Carregando as datas obtidas em uma lista:
datas = []
for d in combinacoes:
  dia = int(d[0])
  mes = int(d[1])  
  datas.append((dia, mes))
```

```python
# Exibindo as datas extraidas:
print(datas)
```

Extraindo outro formato de data:

```python
# Novo texto:
texto2 = "No dia 07/07/1822, foi proclamada a independência do Brasil e no dia 15/11/1899, foi proclamada a república no Brasil"
```

```python
# Novo padrão de data: dd/mm/aaaa
padrao2 = r'\b(\d{1,2})/(\d{1,2})/(\d{4})\b'
```

```python
# Extraindo as datas do texto usando o padrão criado:
combinacoes2 = re.findall(padrao2, texto2)
```

```python
# Carregando as datas em uma nova lista:
novas_datas = []
for d in combinacoes2:
  dia = int(d[0])
  mes = int(d[1])
  ano = int(d[2])
  novas_datas.append((dia, mes, ano))
```

```python
# Exibindo as datas extraidas:
print(novas_datas)
```
