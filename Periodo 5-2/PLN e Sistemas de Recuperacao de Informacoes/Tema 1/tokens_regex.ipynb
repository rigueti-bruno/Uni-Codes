Vamos aplicar a técnica de Tokenização em um texto utilizando Regex. Para isso, iremos selecionar apenas palavras a partir de determinado comprimento.

```python
# Importa a biblioteca de Regex:
import re
```

```python
# Texto simples:
texto = "Este é um texto simples que deve ser tratado."
```

```python
# Uso da função findall da biblioteca Regex para selecionar palavras com comprimento a partir de 4 caracteres:
tokens = re.findall(r'\b\w{4,}', texto)
```

```python
# Exibe as palavras (tokens) selecionados do texto:
print(tokens)
```
