### Análise de Sentimentos como Framework Transformers

Carregando o framework:

```python
from transformers import pipeline
```

Carregando o pipeline de análise de sentimentos:

```python
nlp = pipeline("sentiment-analysis")
```

Texto para análise de sentimentos:

```python
text = "I am really enjoying this movie. It's so much fun!"
```

Realizar a análise de sentimentos:

```python
results = nlp(text)
```

Interpretar o rótulo e a pontuação do sentimento:

```python
sentiment_label = results[0]['label']
sentiment_score = results[0]['score']
```

Imprimir os resultados:

```python
print(f"Texto: {text}")
print(f"Rótulo do Sentimento: {sentiment_label}")
print(f"Pontuação do Sentimento: {sentiment_score:.4f}")
```

Interpretando os sentimentos em um novo texto:

```python
text2 = "I didn't like this movie!"
```

```python
results2 = nlp(text2)
```

```python
sentiment_label2 = results2[0]['label']
sentiment_score2 = results2[0]['score']
```

```python
print(f"Texto: {text2}")
print(f"Rótulo do Sentimento: {sentiment_label2}")
print(f"Pontuação do Sentimento: {sentiment_score2:.4f}")
```
