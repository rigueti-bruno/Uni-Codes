Utilizando o Framework spaCy para Processamento de Linguagem Natual

```python
# Importando o FrameWork:
import spacy
```

```python
# Tokenização:
nlp = spacy.load("pt_core_news_sm") # ativa a função de tokenização
texto = "Eu uso o framework spaCy para me tornar o Batman." # texto a ser tokenizado
doc = nlp(texto) # tokenização do texto
for token in doc:
    print(token.text) # imprime cada token do texto
```

```python
# Marcação de parte do discurso:
for token in doc:
    print(token.text, token.pos_) # imprime cada token e sua parte do discurso
```

```python
# Reconhecimento de entidades nomeadas:
for ent in doc.ents:
    print(ent.text, ent.label_) # imprime cada entidade nomeada e seu rótulo
```
