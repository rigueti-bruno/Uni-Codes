### Exercícios com o Framework spaCy

```python
# Carregando o pacote do Framework:
import spacy
```

```python
# Carregue o modelo spaCy NLP para o português do Brasil
nlp = spacy.load('pt_core_news_sm')
```

```python
# Exemplo de texto para processamento (português)
texto = "No Brasil, existem pessoas amigáveis e inteligentes que gostam de estudar PLN."
```

Processamento do texto usando spaCy:

```python
documento = nlp(texto)
```

Exemplo de Tokenização:

```python
print("Tokens:")
for token in documento:
    print(token.text)
```

Exemplo de Reconhecimento de Parte da Fala:

```python
print("Parte da Fala:")
for token in documento:
    print(token.text, token.pos_)
```

Exemplo de Reconhecimento de Entidade Nomeada:

```python
print("Reconhecimento de Entidade Nomeada:")
for entidade in documento.ents:
    print(entidade.text, entidade.label_)
```
