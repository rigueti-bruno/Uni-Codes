### Exercicio com o Framework Gensim

```python
# Carregando o NLTK e baixando o corpus com as obras de Machado de Assis:
import nltk
nltk.download('punkt')
from nltk.corpus import machado
nltk.download('machado')
```

```python
# Importando o framework Gensom:
import gensim
from gensim.models import Word2Vec
```

```python
# Criando o corpus com as obras de Machado de Assis:
sentences = machado.sents()
```

```python
# Treinamento do modelo Word2Vec:
model = Word2Vec(sentences, vector_size=100,
                 window=5, min_count=1,
                 workers=4)
```

Teste de Semelhança de uma palavra com outras:

```python
word = "amor"
similar_words = model.wv.most_similar(word)
```

```python
# Impressão do resultado:
print(f"Incorporações de Palavras para '{word}':")
for similar_word, similarity in similar_words:
    print(f"{similar_word}: {similarity:.4f}")
```

```python
# Testando outra palavra:
word = "transporte"
similar_words = model.wv.most_similar(word)
```

```python
# Resultados para a palavra "transporte":
print(f"Incorporações de Palavras para '{word}':")
for similar_word, similarity in similar_words:
    print(f"{similar_word}: {similarity:.4f}")
```
