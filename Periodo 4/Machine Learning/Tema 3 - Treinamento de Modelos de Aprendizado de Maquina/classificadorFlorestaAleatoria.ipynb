## Implementando a Floresta Aleatória em Python

1. Importando as Bibliotecas Necessárias:

```python
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.tree import plot_tree
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification
```

2. Definindo os Parâmentros e Treinando o Modelo:

```python
X,y = make_classification(n_samples=1000,n_features=4,n_informative=2,n_redundant=0,random_state=0,shuffle=False) # gerando dados sintéticos
clf = RandomForestClassifier(n_estimators=5,max_depth=2,random_state=0) # definindo os parâmetros do modelo
clf.fit(X,y) # treinando o modelo
print(clf.predict([[0,0,0,0]])) # fazendo uma previsão
```

3. Descobrindo as Importâncias de Características da Floresta Aleatória:

```python
print(clf.feature_importances_)
```

4. Visualizar o Conjunto de Dados Artificialmente Gerado:

```python
pd.concat([pd.DataFrame(X,columns=['atr1','atr2','atr3','atr4']),pd.DataFrame(y,columns=['class'])],axis=1)
```

5. Visualizar as Árvores de Decisão que compõem a Floresta:

```python
for tf in clf.estimators_:
    print("------------------------")
    plot_tree(tf,filled=True)
    plt.show()
```
